```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
data_file = pd.read_csv("movies.csv")
data_file.head()
```

```python
data_file.shape
```

```python
sel_features = ['genres', 'keywords', 'overview', 'cast', 'director']
print(sel_features)
```

```python
for feature in sel_features:
    data_file[feature] = data_file[feature].fillna(' ')
```

```python
main_data = data_file['genres'] + data_file ['keywords'] + data_file ['overview'] + data_file['cast'] +  data_file['director'] 
main_data.head()
```

```python
vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(main_data)
```

```python
print(feature_vectors)
```

```python
similarity = cosine_similarity(feature_vectors)
```

```python
print(similarity)
print(similarity.shape)
```

```python
movie_input = input("Enter Your Favourite Movie Name: ")
```

```python
list_movies = data_file['original_title'].tolist()
print(list_movies)
```

```python
import difflib
find_match = difflib.get_close_matches(movie_input, list_movies)
print(find_match)
```

```python
close_match = find_match[0]
print(close_match)
```

```python
index_input = data_file[data_file.title == close_match]['index'].values[0]
print(index_input)
```

```python
similarity_score = list(enumerate(similarity[index_input]))
print(similarity_score)
```

```python
sorting_similarity = sorted(similarity_score, key = lambda x:x[1], reverse = True)
print(sorting_similarity)
```

```python
print('Movies Suggested: \n')

i = 1
for movie in sorting_similarity:
    index = movie[0]
    title_index = data_file[data_file.index==index]['original_title'].values[0]
    if (i < 10):
        print(i, ".", title_index)
        i=i+1
```
